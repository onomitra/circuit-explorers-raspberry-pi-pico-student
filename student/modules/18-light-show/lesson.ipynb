## Science Spotlight: One Wire, Many Colors

Regular LEDs need one wire each — 10 LEDs = 10 wires. NeoPixels are magic: you can control 100 LEDs with just **ONE wire**, and each can be a different color!

Each NeoPixel has a tiny computer chip inside it that listens for instructions. The data flows down a chain — LED 1 reads its color, then passes the rest of the instructions to LED 2, and so on.

*Your instructor will explain how one wire can control so many LEDs and how each pixel gets 16.7 million possible colors.*

## Meet the NeoPixel Strip

Your NeoPixel strip has **8 LEDs** in a row. Each one contains a tiny **WS2812** chip that can produce any color by mixing **Red**, **Green**, and **Blue** light — just like Module 6's RGB LED, but built into every single pixel!

Each color channel goes from **0** (off) to **255** (maximum brightness):

```
   (255, 0, 0)   = Pure Red
   (0, 255, 0)   = Pure Green
   (0, 0, 255)   = Pure Blue
   (255, 255, 0) = Yellow (Red + Green)
   (255, 0, 255) = Purple (Red + Blue)
   (0, 255, 255) = Cyan (Green + Blue)
   (255, 255, 255) = White (all on)
   (0, 0, 0)     = Off (all off)
```

That's 256 x 256 x 256 = **16,777,216** possible colors per pixel!

The strip has **3 wires** (or pads) on one end:

```
     NeoPixel Strip (8 LEDs)
     ┌───┬───┬───┬───┬───┬───┬───┬───┐
     │ 0 │ 1 │ 2 │ 3 │ 4 │ 5 │ 6 │ 7 │
     └─┬─┴───┴───┴───┴───┴───┴───┴───┘
       │
      DIN  VCC  GND
       │    │    │
     GP15 VBUS  GND
```

- **DIN** = Data In (the ONE signal wire)
- **VCC** = Power (5V from VBUS)
- **GND** = Ground

**Important:** NeoPixels are numbered starting from **0**, not 1! The first LED is `np[0]`, the last is `np[7]`.

## Wiring: NeoPixel Strip

Only 3 wires needed!

### Circuit

```
   Pico GP15  (pin 20) ──── wire ──── NeoPixel DIN (Data In)
   Pico VBUS  (pin 40) ──── wire ──── NeoPixel VCC (5V Power)
   Pico GND   (pin 38) ──── wire ──── NeoPixel GND
```

### Step-by-Step

1. **Look at your NeoPixel strip** — find the end with the labels DIN (or DI), VCC (or 5V), and GND
2. **Connect a wire** from **GP15** on the Pico to the **DIN** pad on the strip
3. **Connect a wire** from **VBUS** on the Pico to the **VCC** pad on the strip
4. **Connect a wire** from **GND** on the Pico to the **GND** pad on the strip

**Double-check:** Make sure power goes to **VBUS** (5V), NOT 3.3V! NeoPixels need 5V to display colors properly.

**Important:** Connect to the **DIN** (Data In) end, not the DOUT (Data Out) end! Data flows in one direction.

## Code Along: Rainbow Cycle

Now for the fun part — **animations!** A rainbow cycle smoothly shifts colors across all the pixels.

We need a helper function that converts a single number (0-255) into a rainbow color. This is called a **color wheel**.

Fill in the blanks:

````micropython
import neopixel
from machine import Pin
import time

np = neopixel.NeoPixel(Pin(15), 8)
NUM_LEDS = 8

def wheel(pos):
    """Convert a position (0-255) to a rainbow RGB color."""
    if pos < 85:
        return (255 - pos * 3, pos * 3, 0)
    elif pos < 170:
        pos -= 85
        return (0, 255 - pos * 3, pos * 3)
    else:
        pos -= 170
        return (pos * 3, 0, 255 - pos * 3)

def rainbow_cycle(wait):
    """Cycle rainbow colors across all pixels."""
    for j in range(_____):
        for i in range(NUM_LEDS):
            pixel_pos = (i * 256 // NUM_LEDS + j) % _____   # What's the max position for the wheel?
            np[i] = wheel(pixel_pos)
        np._____()            # Send colors to the strip!
        time.sleep(wait)

print("Rainbow starting!")
for cycle in range(3):        # Run 3 full rainbow cycles
    rainbow_cycle(0.02)
````

You should see smooth rainbow colors flowing across the strip!

**How it works:**
- The `wheel()` function turns a number (0-255) into a color — 0 is red, 85 is green, 170 is blue, and it wraps back to red
- Each frame, we shift all the colors by a little bit, creating the flowing effect
- The `% 256` keeps the number in range (wraps around)

## Code Along: Comet Animation

A **comet** (or chasing light) effect sends a bright dot across the strip with a fading tail behind it. This looks like a shooting star!

Fill in the blanks:

````micropython
import neopixel
from machine import Pin
import time

np = neopixel.NeoPixel(Pin(15), 8)
NUM_LEDS = 8

def clear():
    for i in range(NUM_LEDS):
        np[i] = (0, 0, 0)
    np.write()

def comet(color, wait):
    """A bright dot with a fading tail chases across the strip."""
    r, g, b = color
    tail_length = 4

    for head in range(NUM_LEDS + tail_length):
        # Clear all pixels first
        for i in range(NUM_LEDS):
            np[i] = (0, 0, 0)

        # Draw the comet head and tail
        for t in range(tail_length):
            pos = head - t
            if 0 <= pos < _____:                  # Only draw if within the strip range
                brightness = 255 - (t * 60)       # Tail gets dimmer
                if brightness < 0:
                    brightness = 0
                scale = brightness / _____         # Divide by max brightness to get a fraction
                np[pos] = (int(r * scale), int(g * scale), int(b * scale))

        np._____()                # Send to strip
        time.sleep(wait)

    clear()

# Run the comet in different colors
print("Comet animation!")
while True:
    comet((0, 0, 255), 0.08)      # Blue comet
    comet((255, 0, 0), 0.08)      # Red comet
    comet((0, 255, 0), 0.08)      # Green comet
````

You should see bright dots with fading tails zooming across the strip in different colors!

**How it works:**
- The "head" of the comet is the brightest pixel
- Behind it, each pixel in the "tail" is dimmer than the one before
- `brightness / 255` gives us a fraction (0.0 to 1.0) to scale the color

## Vocabulary Recap

| Word | What It Means |
|------|---------------|
| **NeoPixel** | A brand name for WS2812 addressable RGB LEDs that can be controlled individually |
| **WS2812** | The chip inside each NeoPixel that receives color data and controls the RGB LED |
| **Addressable LED** | An LED that has its own chip so you can control it individually on a shared wire |
| **RGB** | Red, Green, Blue — the three colors of light mixed together to make any color |
| **Pixel** | One individual LED on the strip, numbered starting from 0 |
| **`np.write()`** | Sends all the color data from the Pico to the strip — nothing changes until you call this |
| **Color wheel** | A function that maps a number (0-255) to a smooth rainbow color |
| **Animation** | Changing colors rapidly in a pattern to create the illusion of movement |
