## Step 1: Brainstorm

Here are some categories to get your ideas flowing. You can pick from these or invent something completely new!

### Smart Home
- **Auto Night Light** — LEDs turn on when the room gets dark (light sensor + LEDs)
- **Temperature Monitor** — OLED screen shows the current temperature with a warning buzzer if it gets too hot
- **Door Alarm** — buzzer goes off when a button is pressed (simulating a door opening)
- **Smart Fan** — motor turns on automatically when temperature rises above a threshold

### Games
- **Reaction Time Tester** — LED lights up randomly, press the button as fast as you can, time is displayed
- **Simon Says** — LEDs flash a pattern, you repeat it with buttons. Each round gets longer!
- **Whack-a-Mole** — NeoPixels light up randomly, press the button when the right one glows
- **Number Guessing** — Pico picks a random number, you guess with the potentiometer, OLED shows hints

### Art & Music
- **Light Show** — NeoPixel animations that react to a potentiometer (speed or color control)
- **Musical Instrument** — buttons play different notes through the buzzer, potentiometer controls pitch
- **Mood Lamp** — RGB LED or NeoPixels slowly cycle through calming colors
- **Sound Visualizer** — NeoPixels respond to button taps with colorful patterns

### Useful Tools
- **Weather Station** — temperature sensor + OLED display, maybe with WiFi to serve a web page
- **Parking Sensor** — ultrasonic distance sensor + buzzer that beeps faster as you get closer
- **Countdown Timer** — set time with potentiometer, OLED shows countdown, buzzer sounds when done
- **Step Counter** — button press counts steps, OLED shows total

### Or Invent Something New!
Combine any sensors, outputs, and ideas you want. The only limit is your imagination (and your component bin).

### Find Your MVP (Minimum Viable Product)

Your MVP is the **simplest working version** of your project. It doesn't have to be perfect or have every feature. It just has to *work*.

For example:
- **Full idea:** Weather station with temperature, humidity, WiFi web page, and alerts
- **MVP:** Temperature sensor shows reading on OLED. That's it. Done. Working.

You can always add features AFTER the MVP works!

## Step 4: Build!

Time to make it real. Here are some tips from experienced makers:

### Start Simple
Build your MVP first. Get ONE thing working before adding the next feature. Don't wire everything at once.

### Test Pieces Separately
Before combining your button, sensor, and display into one big program, test each one on its own. Make sure the button works. Make sure the sensor reads correctly. Then combine them.

### Use Code from Past Modules
You don't have to write everything from scratch! Open your old notebooks and copy the parts you need. That's not cheating — that's how real engineers work.

### Ask for Help
Stuck for more than 5 minutes? Ask your instructor or a classmate. Getting help is a skill, not a weakness.

### It's OK If It's Imperfect
Real engineering projects are never perfect on the first try. If your project mostly works, that's a huge achievement. You can talk about what you'd improve during Demo Day.

```
   The Build Cycle:

   ┌─────────┐     ┌──────┐     ┌──────┐
   │  Wire   │────>│ Code │────>│ Test │
   │one part │     │that  │     │ it!  │
   └─────────┘     │part  │     └──┬───┘
       ^           └──────┘        │
       │                           │
       │     Does it work?         │
       │     No ─── fix it ────────┘
       │     Yes ─── add next part!
       └───────────────────────────┘
```
