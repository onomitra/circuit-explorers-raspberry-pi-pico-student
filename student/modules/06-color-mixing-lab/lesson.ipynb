## Meet the RGB LED

An RGB LED is actually **three tiny LEDs in one package** — one red, one green, one blue. It has **4 legs**:

```
        ┌─────────┐
        │  R G B   │
        │ LED LED LED │
        └─┬─┬─┬─┬─┘
          │ │ │ │
          R │ B G
            │
          GND
        (longest leg)
```

The **longest leg** is the **common cathode** (shared ground). The other three legs control Red, Green, and Blue.

**Important:** The leg order may vary! Check your specific LED. A common pinout is: **Red, GND, Blue, Green** (left to right, with GND being the longest leg).

---
## Wiring the RGB LED

Each color leg needs its own resistor and GPIO pin. The common cathode goes to GND.

### Circuit

```
   Pico GP13 ──── [220Ω] ──── Red leg
                                    │
   Pico GP14 ──── [220Ω] ──── Green leg   ├── RGB LED
                                    │
   Pico GP15 ──── [220Ω] ──── Blue leg
                                    │
   Pico GND  ─────────────── GND leg (longest)
```

### Step-by-Step

1. **Place the RGB LED** on the breadboard — note which leg is longest (GND)
2. **Connect the longest leg** (GND/cathode) to the Pico's **GND**
3. **Connect a 220 ohm resistor** from **GP13** to the **Red** leg
4. **Connect a 220 ohm resistor** from **GP14** to the **Green** leg
5. **Connect a 220 ohm resistor** from **GP15** to the **Blue** leg

**Tip:** If you're not sure which leg is which color, wire them all up and test one at a time with the code below. You can always swap wires!

### Color Mixing Guide

```
   RED + GREEN        = YELLOW
   GREEN + BLUE       = CYAN (light blue)
   RED + BLUE         = MAGENTA (pink/purple)
   RED + GREEN + BLUE = WHITE
```

**Surprised?** Mixing LIGHT is different from mixing PAINT! Red + Green paint makes brown. But red + green LIGHT makes yellow. This is called **additive color mixing**.

## Vocabulary Recap

| Word | What It Means |
|------|---------------|
| **RGB** | Red, Green, Blue — the three primary colors of light |
| **Additive color mixing** | Mixing light — more colors = brighter/whiter (opposite of paint!) |
| **Common cathode** | An RGB LED where all three LEDs share the same ground (GND) leg |
| **Function** | A reusable block of code with a name — like `set_color(r, g, b)` |
| **Parameter** | A value you pass into a function — `r`, `g`, and `b` are parameters |
| **`random.randint()`** | Picks a random whole number between two values |
