## Code Along: Connection Status Checker

Let's build something more polished -- a program that connects to WiFi and keeps checking the connection status.

Fill in the blanks:

````micropython
import network
import time

SSID = "_____"                            # Network name from your instructor
PASSWORD = "_____"                         # Password from your instructor

wlan = network.WLAN(network.STA_IF)
wlan.active(True)
wlan.connect(SSID, PASSWORD)

# Wait for connection with timeout
timeout = 10
while not wlan._____() and timeout > 0:   # What method checks if we're connected?
    print("Connecting... (", timeout, "seconds left)")
    time.sleep(1)
    timeout = timeout - _____              # Decrease by how much each loop?

if wlan.isconnected():
    config = wlan.ifconfig()
    print("\n=== WiFi Connected! ===")
    print("  IP Address :", config[0])
    print("  Subnet Mask:", config[1])
    print("  Gateway    :", config[2])
    print("  DNS Server :", config[3])
    print("========================")
else:
    print("\nFailed to connect! Check your SSID and password.")
````

This is a **robust** connection script -- it doesn't wait forever. If the connection fails, it tells you instead of hanging.

## Challenge: WiFi Signal Strength Monitor

Build a program that connects to WiFi and then **continuously monitors the signal strength** (RSSI).

### Requirements:
- Connect to WiFi
- In a loop, read the current RSSI value and print it
- Show a visual bar made of `#` characters (more `#` = stronger signal)
- Update every 2 seconds

### Useful hints:

```python
# After connecting, get signal strength:
rssi = wlan.status('rssi')            # Returns a negative number like -45

# Typical RSSI values:
#   -30 = amazing (right next to router)
#   -50 = great
#   -70 = okay
#   -90 = very weak

# To make a visual bar:
# Map RSSI from range (-90 to -30) to (0 to 20)
# Then print that many '#' characters
bars = max(0, min(20, (rssi + 90) * 20 // 60))
print("Signal:", '#' * bars, rssi, "dBm")
```

## Vocabulary Recap

| Word | What It Means |
|------|---------------|
| **WiFi** | Wireless technology that uses radio waves to connect devices to a network |
| **SSID** | Service Set Identifier -- the name of a WiFi network (what you see when you search for WiFi) |
| **IP Address** | A unique number assigned to each device on a network, like a home address (e.g. 192.168.1.42) |
| **RSSI** | Received Signal Strength Indicator -- how strong the WiFi signal is (measured in dBm, closer to 0 is stronger) |
| **STA (Station)** | Mode where the Pico connects to an existing WiFi network (like your phone does) |
| **AP (Access Point)** | Mode where the Pico creates its own WiFi network (like a router) |
| **network module** | The MicroPython library that controls WiFi on the Pico 2W |
| **Gateway** | The router's IP address -- the "door" between your local network and the internet |
