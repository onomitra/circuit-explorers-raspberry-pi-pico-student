## Science Spotlight: How Does a Servo Know Where to Point?

A regular motor just spins. A servo motor goes to an exact angle and HOLDS there. That's how robot arms know where to point.

Inside the servo is a motor, some gears, and a position sensor that constantly checks: "Am I at the right angle?" The Pico tells it where to go using a special PWM signal.

*Your instructor will explain how the servo reads PWM pulses and what makes it different from a regular motor.*

### Meet the Servo

The SG90 servo has three wires:
- **Brown (or Black)** = GND (ground)
- **Red** = Power (5V)
- **Orange (or Yellow)** = Signal (this is how the Pico talks to it)

---
## Wiring

### IMPORTANT: The servo needs 5V power, NOT 3.3V!

The Pico's VBUS pin provides 5V directly from the USB cable. This is enough to power the small SG90 servo.

### Servo Wiring

```
   Pico 2W                         SG90 Servo
  ┌─────────┐                     ┌──────────┐
  │         │                     │          │
  │   GP15  ├────────────────────►│ Orange   │  (Signal)
  │         │                     │          │
  │   VBUS  ├────────────────────►│ Red      │  (5V Power)
  │  (5V)   │                     │          │
  │   GND   ├────────────────────►│ Brown    │  (Ground)
  │         │                     │          │
  └─────────┘                     └──────────┘
```

### Potentiometer Wiring (for later)

```
   Pico 2W                        Potentiometer
  ┌─────────┐                     ┌───────┐
  │         │                     │       │
  │   3.3V  ├────────────────────►│ Left  │
  │         │                     │       │
  │  GP26   ├────────────────────►│Middle │  (ADC0)
  │ (ADC0)  │                     │       │
  │   GND   ├────────────────────►│ Right │
  │         │                     │       │
  └─────────┘                     └───────┘
```

### Step-by-Step

1. **Servo signal wire (orange)** connects to **GP15**
2. **Servo power wire (red)** connects to **VBUS** (the 5V pin)
3. **Servo ground wire (brown)** connects to **GND**
4. **Potentiometer left pin** to **3.3V**
5. **Potentiometer middle pin** to **GP26 (ADC0)**
6. **Potentiometer right pin** to **GND**
7. Plug in your USB cable

## Vocabulary Recap

| Word | What It Means |
|------|---------------|
| **Servo motor** | A motor that moves to a specific angle and holds there |
| **SG90** | A small, common hobby servo that moves 0-180 degrees |
| **Pulse width** | How long the PWM signal stays ON each cycle — controls the servo angle |
| **VBUS** | The Pico pin that provides 5V directly from USB power |
| **Mapping** | Converting a value from one range to another (e.g., pot reading to angle) |
| **`duty_u16()`** | Sets the PWM duty cycle using a value from 0 to 65535 |
| **Function** | A reusable block of code with a name — like `set_angle()` |
