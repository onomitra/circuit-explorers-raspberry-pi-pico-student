## Science Spotlight: How Does Electricity Flow?

Electricity is a lot like water flowing through pipes. It needs a complete path вҖ” called a **circuit** вҖ” to flow. If there's a gap anywhere, the flow stops.

We can also control how much electricity flows by putting something in the way вҖ” that's what a **resistor** does.

*Your instructor will explain the water analogy and show you why resistors are so important.*

## Meet Your New Components

Today you'll use four new things. Let's meet them:

### Breadboard
A white plastic board with rows of holes. You push components and wires into the holes to build circuits вҖ” no soldering needed!

### LED (Light Emitting Diode)
A small light with **two legs**:
```
      long leg (+)     short leg (-)
       (anode)          (cathode)
         в”Ӯ                 в”Ӯ
         в”Ӯ     в”Ңв”Җв”Җв”Җв”җ       в”Ӯ
         в””в”Җв”Җв”Җв”Җв”Җв”Ӯ в—Ҹ в”Ӯв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”ҳ
               в””в”Җв”Җв”Җв”ҳ
           (flat side = -)
```
- **Long leg** = positive side (anode) вҖ” electricity goes IN here
- **Short leg** = negative side (cathode) вҖ” electricity comes OUT here
- If you put it in backwards, it won't light up!

### Resistor
A small component with colored bands. It **limits how much electricity flows** through the circuit. Without it, too much electricity would flow and the LED would burn out!

```
    в”Ңв”Җв”Җв”Ө в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳ в”ңв”Җв”Җв”җ
    в”Ӯ   Red Red Brown    в”Ӯ
  leg                   leg
```
Unlike LEDs, resistors work in **either direction** вҖ” you can't put them in backwards.

### Jumper Wires
Colored wires that connect your Pico to the breadboard. Just push them into the holes.

## How a Breadboard Works

The holes in a breadboard are connected **in secret** underneath:

```
   Power Rails          Main Area              Power Rails
   (+)  (-)     a  b  c  d  e    f  g  h  i  j     (+)  (-)
    в”Ӯ    в”Ӯ      в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ    в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ       в”Ӯ    в”Ӯ
    в”Ӯ    в”Ӯ   1 в”Ӯ в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ    в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ в”Ӯ  в”Ӯ    в”Ӯ
    в”Ӯ    в”Ӯ   2 в”Ӯ в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ    в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ в”Ӯ  в”Ӯ    в”Ӯ
    в”Ӯ    в”Ӯ   3 в”Ӯ в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ    в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ в”Ӯ  в”Ӯ    в”Ӯ
    в”Ӯ    в”Ӯ   4 в”Ӯ в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ    в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ  в—Ҹ в”Ӯ  в”Ӯ    в”Ӯ

    в–ј    в–ј     в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ    в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ       в–ј    в–ј
  runs   runs    в—„в”Җв”Җconnectedв”Җв”Җв–ә  в—„в”Җв”Җconnectedв”Җв”Җв–ә  runs   runs
  down   down    across the row   across the row   down   down
```

**Key rules:**
- Each **row** (numbered) is connected across вҖ” holes a-b-c-d-e are all linked
- The **center gap** breaks the connection вҖ” left side and right side are separate
- The **power rails** (+/-) on the edges run up and down the whole board
- To connect two components, put them in the **same row**

## Wiring Your Circuit

Follow these steps carefully. **Don't plug in the USB cable yet!**

### Step-by-step:

**Step 1:** Place the **LED** into the breadboard
- Long leg (+) in row 5, column e
- Short leg (-) in row 6, column e

**Step 2:** Place the **resistor**
- One leg in row 5, column a (same row as the LED's long leg)
- Other leg in row 3, column a

**Step 3:** Connect a **jumper wire** from **GP15** on the Pico to row 3 on the breadboard

**Step 4:** Connect a **jumper wire** from **GND** on the Pico to row 6 on the breadboard (same row as LED's short leg)

```
    Pico                  Breadboard
   GP15 в”Җв”Җв”Җв”Җ wire в”Җв”Җв”Җв”Җв”Җв”Җ row 3, col a
                              в”Ӯ
                         [RESISTOR] (row 3 to row 5)
                              в”Ӯ
                         LED long leg (+)  row 5
                         LED short leg (-) row 6
                              в”Ӯ
    GND в”Җв”Җв”Җв”Җ wire в”Җв”Җв”Җв”Җв”Җв”Җ row 6, col a
```

### Before you plug in, double-check:
- [ ] LED long leg and resistor are in the **same row** (row 5)
- [ ] LED short leg and GND wire are in the **same row** (row 6)
- [ ] Wires go to **GP15** and **GND** on the Pico

Now plug in your USB cable!

## Vocabulary Recap

| Word | What It Means |
|------|---------------|
| **Circuit** | A complete path that electricity flows through вҖ” it must loop back to where it started |
| **Breadboard** | A board with connected holes for building circuits without soldering |
| **Resistor** | A component that limits how much electricity flows вҖ” protects other components |
| **Anode** | The positive (+) side of an LED вҖ” the long leg |
| **Cathode** | The negative (-) side of an LED вҖ” the short leg |
| **Voltage** | The "push" that makes electricity flow (measured in Volts) |
| **Current** | How much electricity is flowing (measured in Amps) |
| **Resistance** | How much something slows down the flow of electricity (measured in Ohms) |
