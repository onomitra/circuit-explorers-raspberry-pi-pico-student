## Science Spotlight: What is a Microcontroller?

Your phone or laptop is a full computer — it runs apps, plays videos, and does a million things at once.

The Pico is something different. It's a **microcontroller** — a mini-brain designed to do **one job really well**. Think of the things around you that don't need a full computer: a traffic light, a microwave timer, a motion-sensor light. Those all use microcontrollers!

## Step 2: Flash MicroPython Firmware onto Your Pico

Before Thonny can talk to your Pico, you need to install **MicroPython** on it. You only do this once.

1. **Download the MicroPython firmware** — go to [raspberrypi.com/documentation/microcontrollers/micropython.html](https://www.raspberrypi.com/documentation/microcontrollers/micropython.html) and download the `.uf2` file for the **Raspberry Pi Pico 2W** (make sure it says Pico 2W, not just Pico)

2. **Hold the BOOTSEL button** on your Pico — it's the small white button near the USB port

3. **While holding BOOTSEL**, plug the USB cable into your Pico, then into your computer

4. **Let go of BOOTSEL** — your Pico should now appear as a USB drive on your computer called `RPI-RP2`

```
    Your Computer                    Pico 2W
   ┌──────────┐      USB Cable     ┌──────────┐
   │          │ ══════════════════> │ BOOTSEL  │
   │          │                     │  held!   │
   └──────────┘                     └──────────┘
         │
         ▼
   Pico appears as
   a USB drive: RPI-RP2
```

5. **Drag and drop** the `.uf2` file you downloaded onto the `RPI-RP2` drive — just like copying a file

6. The Pico will **automatically reboot** and the drive will disappear — that means it worked!

> **Why does it reboot?** Copying the `.uf2` file installs MicroPython as the Pico's operating system. Once installed, it boots straight into MicroPython every time you plug it in.

## Step 4: Your First Line of Code

Let's make sure everything works. In the **Shell** at the bottom of Thonny (where you see `>>>`), type this and press Enter:

## Step 4: Meet the Onboard LED

Your Pico has a tiny **LED** (Light Emitting Diode) built right into the board. You don't need any wires to use it!

We control it through something called **GPIO** (General Purpose Input/Output) — these are the Pico's connections to the outside world. The onboard LED is connected to a special pin called `"LED"`.

Let's turn it on!

1. Turn the LED **on**
2. **Wait** a moment

4. **Wait** again
5. **Repeat** forever!

## Challenge: SOS!

In Morse code, **SOS** is the international distress signal:

```
S = ● ● ●       (3 short blinks)
O = ▬ ▬ ▬       (3 long blinks)
S = ● ● ●       (3 short blinks)
```

Can you program your Pico to blink SOS? Think about:
- How long is a "short" blink? How long is a "long" blink?
- How long do you wait between the S and the O?
- How do you make it repeat?

Ask your instructor for help if you get stuck!

## Vocabulary Recap

| Word | What It Means |
|------|---------------|
| **Microcontroller** | A tiny computer designed to do one specific job |
| **GPIO** | General Purpose Input/Output — the Pico's pins that connect to the outside world |
| **LED** | Light Emitting Diode — a small light |
| **MicroPython** | A version of the Python programming language made for small devices like the Pico |
| **REPL** | Read-Eval-Print Loop — the Shell in Thonny where you type commands one at a time |
| **import** | A command that loads extra tools (modules) for your code to use |
| **loop** | Code that repeats — `while True` repeats forever, `for` repeats a set number of times |
