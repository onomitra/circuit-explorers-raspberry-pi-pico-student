## Science Spotlight: Input vs Output

Until now, your Pico has only been giving orders — "turn on," "turn off," "blink." That's **output**.

But what if it could listen? A button is one of the simplest ways for **you** to talk to your Pico. When you press it, you're sending it a signal — that's **input**.

*Your instructor will explain how the Pico reads a button press, and why we need something called a "pull-up resistor" to make it work reliably.*

## Meet the Push Button

A push button is a simple switch. Press it down and it **connects** two points. Let go and the connection **breaks**.

Your button has **4 pins**, but they're connected in pairs:

```
        ┌─────────┐
        │ ┌─────┐ │
   A ───┤ │     │ ├─── B
        │ │PRESS│ │
   A ───┤ │     │ ├─── B
        │ └─────┘ │
        └─────────┘

   A-A are ALWAYS connected (left pair)
   B-B are ALWAYS connected (right pair)
   A-B are connected ONLY when pressed
```

**Tip:** Place the button so it **bridges the center gap** of your breadboard. This way, each pair of pins is on a separate side.

## Wiring: Button Only

Let's start with just the button — no LED yet. We'll add that later.

### Circuit

```
   Pico GP14 ──── wire ──── Button Pin A
                                  │
                            [press to connect]
                                  │
   Pico GND  ──── wire ──── Button Pin B
```

### Step-by-Step

1. **Place the button** across the center gap of the breadboard
2. **Connect a wire** from **GP14** on the Pico to one side of the button (Pin A)
3. **Connect a wire** from **GND** on the Pico to the other side of the button (Pin B)

That's it — just 2 wires! We don't need an external resistor because we'll use the Pico's **built-in pull-up resistor** in our code.

## Understanding the Values

You probably noticed something that feels **backwards**:

| Button State | Value |
|-------------|-------|
| **Not pressed** | `1` |
| **Pressed** | `0` |

Why is "pressed" = 0? It seems like it should be the other way around!

This is because of the **pull-up resistor**. It gently "pulls" the pin up to `1` when nothing is pressing the button. When you press the button, it connects the pin to **GND** (ground), which forces it down to `0`.

*Your instructor can explain more about why this works so reliably.*

**The key rule:** With `Pin.PULL_UP`, check for `0` to detect a press.

## Wiring: Add the LED

Now let's add an LED so the button can control something! Keep your button wired and add:

```
   Button Circuit (from before):
   GP14 ──── Button ──── GND

   LED Circuit (same as Module 2):
   GP15 ──── [220Ω Resistor] ──── LED (+) ──── LED (-) ──── GND
```

### Step-by-Step

1. **Keep** the button wired to GP14 and GND
2. **Add an LED** to the breadboard (long leg = +)
3. **Add a 220Ω resistor** connecting GP15's wire row to the LED's long leg row
4. **Connect** the LED's short leg row to GND

Both the button and LED can share the same GND connection on the Pico.

## Code Along: Mini Table Lamp (Toggle)

A doorbell is nice, but a **table lamp** is better — press once to turn ON, press again to turn OFF.

This is called a **toggle**. We need to remember whether the LED is currently on or off.

Fill in the blanks:

````micropython
from machine import Pin
import time

button = Pin(14, Pin.IN, Pin.PULL_UP)
led = Pin(15, Pin.OUT)

led_on = _____                    # Start with LED off. True or False?

while True:
    if button.value() == 0:       # Button is pressed
        led_on = not led_on       # Flip: True becomes False, False becomes True
        led.value(_____)          # Use led_on to set the LED. What goes here?
        time.sleep(_____)         # Wait a moment. Why? Ask your instructor!
````

Press the button once — LED turns on. Press again — LED turns off. You built a lamp!

**Why the `time.sleep()`?** Without it, the loop runs thousands of times per second. One quick button press would toggle the LED hundreds of times! The sleep gives your finger time to release the button. This is a simple form of **debouncing**.

## Vocabulary Recap

| Word | What It Means |
|------|---------------|
| **Input** | Information coming IN to the Pico from the world (like a button press) |
| **Output** | Information going OUT from the Pico to the world (like turning on an LED) |
| **Pull-up resistor** | A resistor that gently holds an input pin at `1` until something (like a button) pulls it to `0` |
| **Debounce** | Ignoring the rapid on/off signals caused by a button's mechanical bounce |
| **Conditional (if/else)** | Code that makes a decision — do one thing IF something is true, ELSE do something different |
| **Toggle** | Switching between two states — on becomes off, off becomes on |
| **Boolean** | A value that is either `True` or `False` — used to track states like "is the LED on?" |
