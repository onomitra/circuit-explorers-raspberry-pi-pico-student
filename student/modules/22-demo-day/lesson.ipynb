## Your Presentation Guide

You'll have **2-3 minutes** to present your project to the class. Here's what to cover:

### 1. Project Name
What did you call it?

### 2. What Does It Do?
Explain in 1-2 sentences. Then **show it working live!**

### 3. Components Used
What sensors, outputs, and other parts did you use? Hold them up or point them out on your breadboard.

### 4. The Hardest Part
What gave you the most trouble? How did you solve it (or work around it)?

### 5. What You'd Add
If you had more time, what feature would you add next?

### 6. One Surprising Thing You Learned
What's something that surprised you during this course or during your project?

---

### Presentation Prep Checklist

Before your turn, make sure:

- [ ] Your project is plugged in and ready to demo
- [ ] You've tested it at least once today -- it works!
- [ ] You know how to start your code in Thonny
- [ ] You can explain what it does in plain language
- [ ] You know which components you used and which pins they're on

**Tip:** If your project has a warm-up time (like the PIR sensor), start it a few minutes before your turn!

## Giving Good Feedback

When watching a classmate's presentation, your job is to be a **supportive audience**. Use the feedback cards your instructor handed out.

### What Good Feedback Looks Like

**One thing you liked:**
- "I thought it was cool that your buzzer played different tones for different temperatures."
- "Your NeoPixel animation was really smooth."
- "I liked how you explained the voltage divider -- that helped me understand it better."

**One question you have:**
- "How did you figure out the right threshold for the light sensor?"
- "What would happen if you added a second button?"
- "Did you try connecting it to WiFi?"

### What to Avoid
- Don't say "that's easy" -- every project took real effort
- Don't point out bugs unless asked -- today is about celebrating, not debugging
- Don't compare projects -- each one is unique

**Remember:** Everyone in this room started from zero. Every working project is an achievement.

## What's Next? Keep Building!

This course is over, but your journey with electronics and programming is just beginning. Here are some great resources to keep learning:

### Freenove Tutorials
Your Pico kit came from Freenove. They have tons of additional projects and tutorials:
- Visit [freenove.com](https://freenove.com) and look for Raspberry Pi Pico projects
- Many projects go beyond what we covered -- motor drivers, LCD screens, IR remotes, and more

### MicroPython Documentation
The official MicroPython docs explain every feature of the language:
- [docs.micropython.org](https://docs.micropython.org)
- The "Quick Reference for RP2" page is especially useful for Pico

### Raspberry Pi Foundation
The people who make the Pico have free learning resources:
- [projects.raspberrypi.org](https://projects.raspberrypi.org)
- Great step-by-step projects for all skill levels

### Arduino
Arduino is another popular platform for electronics projects. The concepts you learned here (GPIO, PWM, ADC, sensors, actuators) apply directly to Arduino too:
- [arduino.cc/en/Tutorial](https://www.arduino.cc/en/Tutorial/HomePage)
- Arduino uses C++ instead of Python, but the hardware concepts are the same

### YouTube Channels
Search for "Raspberry Pi Pico projects" or "MicroPython tutorials" on YouTube. Channels like **DroneBot Workshop**, **ExplainingComputers**, and **Kevin McAleer** have great Pico content.

---

**The most important thing:** Keep building things. Every project you build -- even the ones that don't work perfectly -- teaches you something new. The best way to learn electronics is to DO electronics.
