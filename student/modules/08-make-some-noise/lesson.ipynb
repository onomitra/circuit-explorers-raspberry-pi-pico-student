## Science Spotlight: Sound and Frequency

Sound is just air vibrating really fast. Make the air vibrate **440 times per second** and you hear the musical note A. Make it vibrate **523 times per second** and you hear a C.

Higher number = higher pitch. Lower number = lower pitch.

*Your instructor will explain how the Pico can create different notes by vibrating a buzzer at precise frequencies — and why we use PWM to do it.*

---
## Meet the Passive Buzzer

There are two types of buzzers:

- **Active buzzer:** Has a built-in oscillator. Give it power and it makes ONE fixed tone. Boring!
- **Passive buzzer:** No built-in oscillator. YOU control the frequency, so you can play ANY note. Fun!

We're using the **passive buzzer** from your Freenove kit.

```
    +--------+
    | BUZZER |
    |  )))   |
    +--------+
     +     -
     |     |
   GP15   GND
```

The buzzer has 2 pins:
- **(+)** marked with a plus sign or longer leg — connects to GP15
- **(-)** shorter leg — connects to GND

**How to tell them apart:** Look for a `+` marking on top of the buzzer, or the longer leg is positive (just like an LED).

## Wiring: Buzzer + Button

We'll wire both the buzzer (output) and a button (input) so we can build a doorbell later.

### Circuit

```
   Buzzer Circuit:
   Pico GP15 ──── wire ──── Buzzer (+)
                             Buzzer (-) ──── GND

   Button Circuit (same as Module 4):
   Pico GP14 ──── wire ──── Button (one side)
   Pico GND  ──── wire ──── Button (other side)
```

### Step-by-Step

1. **Place the buzzer** on the breadboard — note which leg is (+) and which is (-)
2. **Connect a wire** from **GP15** on the Pico to the buzzer's **(+)** pin
3. **Connect** the buzzer's **(-)** pin to **GND** on the Pico
4. **Place the button** across the center gap of the breadboard
5. **Connect a wire** from **GP14** to one side of the button
6. **Connect a wire** from **GND** to the other side of the button

Plug in your USB cable.

## Code Along: Doorbell

Let's combine the buzzer with the button to make a doorbell! Press the button and the buzzer sounds.

Fill in the blanks:

````micropython
from machine import Pin, PWM
import time

buzzer = PWM(Pin(15))
button = Pin(_____, Pin.IN, Pin._____)    # Which pin? What pull mode?

while True:
    if button.value() == _____:           # What value means "pressed"?
        buzzer.freq(800)
        buzzer.duty_u16(32768)            # Buzzer ON
    else:
        buzzer.duty_u16(_____)            # Buzzer OFF — what duty = silence?
    time.sleep(0.01)
````

Press and hold the button — you should hear a tone. Release it — silence.

## Code Along: Play a Melody

A scale is nice, but music needs **rhythm** — some notes are long and some are short.

We'll store each note as a **tuple**: `(frequency, duration)`. A frequency of **0** means a rest (silence).

Here's "Mary Had a Little Lamb". Fill in the blanks:

````micropython
from machine import Pin, PWM
import time

buzzer = PWM(Pin(15))

# Mary Had a Little Lamb
# Each tuple is (frequency, duration_in_seconds)
melody = [
    (330, 0.4), (294, 0.4), (262, 0.4), (294, 0.4),
    (330, 0.4), (330, 0.4), (330, 0.8),
    (294, 0.4), (294, 0.4), (294, 0.8),
    (330, 0.4), (392, 0.4), (392, 0.8),
    (330, 0.4), (294, 0.4), (262, 0.4), (294, 0.4),
    (330, 0.4), (330, 0.4), (330, 0.4), (330, 0.4),
    (294, 0.4), (294, 0.4), (330, 0.4), (294, 0.4),
    (262, 0.8)
]

for freq, duration in _____:          # What do we loop through?
    if freq == 0:                     # Rest (silence)
        buzzer.duty_u16(_____)
    else:
        buzzer.freq(freq)
        buzzer.duty_u16(_____)
    time.sleep(duration)
    buzzer.duty_u16(0)                # Brief gap between notes
    time.sleep(0.05)

buzzer.duty_u16(0)
````

Can you recognize the tune? Your Pico just played its first song!

Notice how some notes have a duration of `0.8` (long) and others `0.4` (short) — that's what gives the melody its rhythm.

## Vocabulary Recap

| Word | What It Means |
|------|---------------|
| **Frequency** | How many vibrations per second — measured in Hertz (Hz). Higher frequency = higher pitch |
| **Hertz (Hz)** | The unit of frequency — 440 Hz means 440 vibrations per second |
| **Passive buzzer** | A buzzer with no built-in oscillator — you control the frequency to play any note |
| **Active buzzer** | A buzzer with a built-in oscillator — makes only one fixed tone |
| **Duty cycle** | The percentage of time the signal is ON — for buzzers, 50% (32768) works best |
| **Tuple** | A pair (or group) of values stored together, like `(440, 0.5)` for a note and its duration |
| **Melody** | A sequence of notes played in order with specific timing |
